<h3>1.1 Introduction to the Seaborn Package</h3>

<big>Seaborn is a Python statistical data visualization based library which is used for making very informative, attractive plots via various functionalities. We all know that the ability to communicate insights discovered from a particular dataset via visualizations is very imperative in data science, because most clients more often than not rely on visual cues rather than the complexities of a machine learning model. Seaborn does this job perfectly well, combining various tools and techniques to create visuals to demystify the underlying patterns that are inherent in data. Seaborn is built on top of matplotlib (which i assume we all know about) and works seamlessly with pandas data structures i.e Dataframes, Series e.t.c. Seaborn provides various functionalities for visualizing the relationships between features in a dataset, univariate and bi-variate distributions, tools for choosing styles and color palettes to suit your preferences, high-level abstraction to visualize complex datasets. Subsequently, in this notebook, we shall explore practically most of the concepts explained here</big>


<h3>1.2 Seaborn VS. Matplotlib</h3>
<big><p>As aforementioned, Seaborn was built on the Matplotlib library and is well-suited for statistical data visualization primarily. However, the Seaborn library significantly extends the functionality of the Matplotlib library. It most importantly provides more beautiful plots than matplotlib with a better aesthetic appeal which is very vital to effective data visualization. It also uses fewer syntax which reduces coding time. Another very apparent issue is that matplotlib does not quite well integrate smoothly with pandas dataframes which could be very frustrating since Pandas is the most common library for exploratory data analysis.</p>

<p>Nevertheless, good knowledge of Matplotlib would always come in handy in understanding the intricacies of the seaborn library.</p> 

<p>Other libraries used for data visualization include Plotly, d3.js, Bokeh, ggplot, Geoplotlib, Pygal, Altair e.t.c.</p></big>

<h3>1.3 Practical Data Visualization with the Financial Inclusion Dataset (using seaborn)</h3>

Overview :- The financial inclusion dataset is a zindi-based dataset which aims to create a machine learning model to predict which individuals are most likely to have a bank account or not based on several demographic and financial service features. The variable definitions file is also available in this folder in CSV format. Please, pay attention as we explore this dataset visually 

```python
# STEP1 - IMPORT THE REQUIRED LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
import seaborn as sns #The seaborn library 
%matplotlib inline 
```

```python
print(plt.style.available)
```

```python
#TO SET YOUR CUSTOM PREFERRED COLOR PALETTE

pref_colors = ['crimson', 'blue', 'green','purple']
sns.palplot(sns.color_palette(pref_colors))

sns.set_palette(sns.color_palette(pref_colors))

#TO SET PREFERRED STYLE
sns.set(style="ticks", color_codes=True)
sns.set_style('white', {'axes.axisbelow':True})
```

```python
data = pd.read_csv('bank_accounts.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
categorical_features = data.select_dtypes(include=[np.object]).columns
numerical_features = data.select_dtypes(include=[np.number]).columns
```

```python
print(categorical_features,numerical_features)
```

Working with the categorical, discrete features - for Univariate and Bivariate visualization using sns.catplot(kind = "")

## Categorical Estimate plots

#### Count plot

```python
#create the plot
count_plot = sns.catplot(x="education_level", kind="count",palette=sns.color_palette(pref_colors),data=data)
#set the title
plt.title("Count plot")
plt.xticks(rotation=90)
count_plot.savefig("countplot.png")
```

```python
#Using Hue 
count_plot = sns.catplot(x="education_level", kind="count",palette=sns.color_palette(pref_colors),data=data, hue="bank_account")
#set the title
plt.title("Count plot")
plt.xticks(rotation=90)
```

#### Barplot

```python
#creating the bar-plot
sns.catplot(x="relationship_with_head", y="household_size", data=data, kind="bar")
plt.xticks(rotation=90)
```

```python
#Using Hue
sns.catplot(x="relationship_with_head", y="household_size", data=data, kind="bar",hue="bank_account")
plt.xticks(rotation=90)
```

#### Point Plots

```python
sns.catplot(x="bank_account", y="age_of_respondent",data=data, hue="gender_of_respondent",kind='point')
```

```python
#Changing marker and linestyles for more attractiveness
sns.catplot(x="relationship_with_head", y="age_of_respondent",data=data, hue="gender_of_respondent",kind='point', palette={'Male':"g","Female":"m"}, markers=["^","o"], 
               linestyles=["-","--"])
plt.xticks(rotation= 90)
```

#### Scatter Plots

```python
sns.catplot(x="relationship_with_head",y="age_of_respondent", data=data)
plt.xticks(rotation=90)
```

#### Boxplots

```python
fig, ax = plt.subplots(1,2, figsize=(18,8))

sns.boxplot(x="bank_account", y="age_of_respondent", data=data, ax=ax[0])
sns.boxplot(x="bank_account", y="age_of_respondent",data=data, ax=ax[1], hue= "gender_of_respondent", dodge=True)
ax[0].set_title("Without Hue")
ax[1].set_title("With Hue")
```

#### BoxenPlot - Very similar to Boxplots however and well suited for large datasets

```python
sns.catplot(x="bank_account",y="household_size",kind="boxen", data=data)
```

#### Violin Plot - A Violin plot combines a boxplot with a kernel density estimation procedure which will be explained in obtaining distribution plots for continuous variables i.e

```python
fig, ax = plt.subplots(1,2,figsize=(18,6))
sns.violinplot(x="bank_account", y="age_of_respondent", ax=ax[0],split=True, data=data)
sns.violinplot(x="bank_account",y="age_of_respondent", ax=ax[1],aplit=True,data=data, hue="gender_of_respondent")
```

#### Visualizing Higher dimensional relationships using FacetGrids
<p>A FacetGrid forms a matrix of panels defined by row and column by dividing the
variables. Due of panels, a single plot looks like m ultiple plots. It is very helpful to
analyze all combinations in two discrete variables.</p>

```python
facet = sns.FacetGrid(data = data,col="bank_account", row="location_type")
facet.map(sns.distplot, "household_size",color="red")
plt.tight_layout(w_pad = 3)
```

#### Working with PairGrids

Similar to pairplot explained below but then it gives you better control over the kind of plot you want for each grid

```python
#Instantiate the pair grid passing the data as an argument
grid = sns.PairGrid(data)
#Map the specific plot-kinds you want to any area of the grid
grid.map_diag(sns.distplot)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
```

```python
sns.catplot(x="bank_account", y="age_of_respondent",row="gender_of_respondent",kind="box",data=data)
```

#### Working with Continuous Variables - Visualizing distributions for continuous features

```python
#creating the distplot for univariate analysis of the age_of_respondent feature
sns.distplot(data["age_of_respondent"], bins=10)
```

```python
#To remove the Kernel Density Estimation Plot - the blue line in the previous plot
sns.distplot(data["age_of_respondent"], bins=10, kde=False,color="purple", hist_kws={"edgecolor":"black","linewidth":2})
```

#### Joint Plot for Bivariate Data

```python
#To create the plot 
new_df = pd.DataFrame(np.random.randint(0,100,size=(100,2)), columns=['X','Y'])
sns.jointplot(x='X',y='Y',data=new_df, kind="reg")
```

```python
#Joint plot using the scatter plot for the bivariate distribution
sns.jointplot(x='X',y='Y',data=new_df, kind="scatter")
```

#### Scatter Plot

```python
sns.scatterplot(x='X', y='Y', data=new_df)
```

#### PAIRPLOT - basically, a pairplot visualizes multiple pairwise relationships for each pair of columns in a dataset or dataframe. It performs both univariate nad bivariate distribution plots of each of those columns

```python
#create the plot using the sns.pairplot() method
sns.pairplot(data)
```

```python
#using hue with the pairplot
sns.pairplot(data, hue="bank_account", diag_kind="hist")
```

#### Plots for regression analysis - Linear model plots (lmplot())

```python
sns.lmplot(x="X", y="Y", data=new_df)
```

#### Working with Heatmaps 
<p>Heatmaps are used to show the correlation between numerical features in a dataset. A seaborn heatmap plot can be used to effectively visualize this relationship in very little lines of code.</p>

```python
#Create the dataset showing the correlation of the numerical features using the .corr() method
corr_data = data.corr()
corr_data
```

```python
#Create the plot of the heatmap using seaborn
sns.heatmap(corr_data, annot=True,cmap="RdYlGn",linewidth=0.2, annot_kws={'size':15})
fig=plt.gcf()
fig.set_size_inches(15,10)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15, rotation=90)
```

#### Working with clustermaps - a variant of heatmaps which produces hierarchical clusters of the features in the heatmaps

```python
sns.clustermap(corr_data, annot=True)
```
