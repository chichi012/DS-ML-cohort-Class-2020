### 1.1 What is Ensemble Learning ?

<p> We all understand that Predictive models are at the core of Machine learning, and as Data scientists we are very concerned about improving the accuracy and also other metrics of performance to very satisfiable levels. Ensemble learning is a very essential topic in Machine learning which discusses how to do that just perfectly. In the simplest of terms, The whole idea behind Ensemble learning is to provide a very potent technique to improve the performance of our machine learning models by strategically combining multiple different models using different algorithms with different performance metrics to form a <i>"Super-Model"</i>. So basically, when different models are combined together strategically, it produces better, accurate and more robust predictions than predictions by an individual model. </p>

<img src="Image2_Ensemble_cndjda.png" width="500px" height="500px">

<p> The image above demystifies well enough the major concept behind Ensemble learning. Training data is passed into different models with different predictions and those models are combined via a meta-learner to give a better model </p>.



### 1.2 Why Ensemble Learning ?

<p> The primary goal of Ensemble methods in machine learning is to make better predictions by combining different models. Ensemble methods usually perform well and most times provide winning solutions in machine learning competitions on different platforms - Kaggle, Zindi, MachineHack. Hence their superiority to other individual models is why Ensemble learning is so important. Ensemble learning also proves effective in the bias-variance tradeoff helping to produce models void of overfitting or underfitting </p>

### 1.3.1 Max Voting :- 
<p>The max voting method is generally used for classification problems. In this technique, multiple models are used to make predictions for each data point. The predictions by each model are considered as a ‘vote’. The predictions which we get from the majority of the models are used as the final prediction. The infographic below can help to build a visual intuition behind this concept</p>
<img src="Image5_Ensemble_jchh9g.png">

### 1.3.2 Averaging Methods :-
In this method basically, the mathematical average of the predictions of the models combined is taken to form the ensemble model predictions. It is majorly used in regression problems or calculating the probabilities for classification problems. The weighted average method is also another kind of averaging method which assigns different weights or importance to different models to be ensembled based on their potency in solving the problem

### 1.3.3 Stacking :-
<img src="stacking2.png" width="500px" height="500px">
<p> Stacking (also known as Stacked Generalization) uses a different paradigm from the aforementioned averaging ang voting methods. This method combines intermediate predictions from different individual models and those predictions are sent in to a final model (meta-learner) which often produces better predictions. The training set is first split into K-folds then , then the base model is fitted on K-1 parts and predictions are made on the Kth part in the train set. The predictions from each base model are then used as features for the second-level model which makes predictions on the test set. Blending is also similar to stacking</p>

### 1.3.4 Bagging (Bootstrap Aggregation) :-
<p> Bagging is a very effective ensemble method which improves the performance of single models and reduces overfitting by combining the result of different models modelled on different sub-samples of the training data. The basic process involved is initially, Multiple sub-sets of the data are produced and base models (weak learners) are fitted on the sub-sets. These base models are then combined in parallel independent of each other to give a more robust prediction with reduced variance and over-fitting. The picture above gives a very infographic explanation of this concept :-</p>

<img src="Screenshot-from-2018-05-08-13-11-49-768x580.png" width="400px" height="400px">

Common bagging algorithms include bagging meta-estimator, and yeah the popularly used Random Forest


### 1.3.5 Boosting :-
<p> Boosting basically combines weak learners into strong ones and helps to effectively reduce bias and variance in our model. The major difference between boosting and bagging is that boosting trains the base models sequentially while bagging trains them in parallel. Each subsequent model tries to correct the errors of the previous model. </p>

### 1.4 Hyperparameter Tuning :-

<p>Machine learning models have both parameters and hyperparameters. A model parameter is like an configured variable that is inherent to the machine learnig model and whose value can be estimated from data. The model parameters help in making predictions i.e The slope, intercept of a Linear Regression Model are model parameters and they cannot be tuned or set manually. However, in contrast to model parameters - A model Hyperparameter are configuration variables that help the machine learning model to learn and are external to the model. Hyperparameters include the k in k-nearest neighbors, the n-estimators in RandomForest, the regularization parameter in Logistic Regression, The learning rate in neural network models e.t.c. As data scientist, still on the quest for a very accurate model, it is expedient that we tune some of the hyperparameters in our model to figure out the one that provides the best ad most suitable model for the specific data. Hyperparameter tuning hence is a means of choosing the most optimal hyperparameters for a machine learning model. Hyperparameter tuning can help to combat overfitting and reduce variance in models.</p>


### 1.4.1 Methods for Hyperparameter Tuning  :-

<ul>
    <li>Grid Search </li>
    <li>Randomized Search</li>
</ul>

<p>Grid search helps to optimally tune hyperparameters by searching exhaustively through a specified subset of hyperparameters of different values. So the gridsearch combines every possible combination of hyperparameters and selects the one which gives the best performance after iterative-testing. This algorithm is implemented using the sklearn.model_selection.GridSearchCV library. RandomizedGridSearch in constrast to the GridSearchCV searches through the subset of hyperparameters randomly rather than exhaustively and is usually faster. This can also be implemented using the sklearn.model_selection.RandomizedGridSearch </p>

<p>We are using the IBM HR Attrition dataset for this practical session, the dataset was downloaded from kaggle. <a href="https://www.kaggle.com/pavansubhasht/ibm-hr-analytics-attrition-dataset/data">Link Here for download and better Description of the dataset</a></p>

Import required libraries and get started

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import scipy
import warnings
warnings.filterwarnings("ignore")

```

```python
data = pd.read_csv("EmployeeAttrition.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()

#Clean Data
```

```python
target = data.Attrition
target.value_counts()
```

```python
data['Attrition'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%', shadow=True)
```

```python
data.Attrition = data.Attrition.map({'Yes':1, 'No':0})
```

```python
cat_features = data.select_dtypes(include = [np.object]).columns
num_features = data.select_dtypes(exclude = [np.object]).columns
print(cat_features)
print(num_features)
```

```python
for i in cat_features:
    print(data[i].value_counts())
```

```python
#constant_features - Features with a constant value in the dataset. We need to drop them because they are redundant to our model

constant_feats = ['Over18', 'EmployeeCount', 'StandardHours']
data.drop(constant_feats, inplace =True, axis=1)
```

```python
#Categorical Feature Encoding 
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
for i in ['OverTime', 'Gender']:
    data[i] = lbl.fit_transform(data[i])
    
    
```

```python
data_encoded = pd.get_dummies(data, drop_first = True)
```

```python
data_encoded.shape
```

```python
data_encoded.sample()
```

```python
#Lets view the correlation between features using a seaborn Heatmap
#Can we detect any Multi-colinear features ?
plt.figure(figsize = (24,24))
sns.heatmap(data_encoded.corr(), annot=True)
```

```python
corr_matrix = data_encoded.corr().abs()

upper= corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

print(upper)
```

```python
columns_to_drop = [col for col in upper.columns if any(upper[col] > 0.8)]
print(columns_to_drop)
data_encoded.drop(columns_to_drop, inplace=True, axis=1)
```

```python
target = data_encoded.Attrition.astype(int)
data_encoded.drop("Attrition", axis=1, inplace =True)

# #feature selection
# from sklearn.feature_selection import SelectKBest, chi2
# fs = SelectKBest(score_func =chi2, k =40)

# #apply feature_selection
# data_selected = fs.fit_transform(data_encoded, target)
```

```python
#Scaling the Data
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, Normalizer

std = StandardScaler()
data_scaled = std.fit_transform(data_encoded)
```

```python
data_scaled
```

### Predictive Modelling

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
x_train, x_test, y_train, y_test = train_test_split(data_scaled, target, stratify=target, random_state=42, test_size= 0.3)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
#Using Individual Algorithms to see what our baseline performance is :-

#Importing the required Libraries
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# Some Ensemble Model Libraries
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.ensemble import VotingClassifier, BaggingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score,classification_report,recall_score,f1_score,precision_score,confusion_matrix
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier


```

```python
models = []

models.append(('RFC', RandomForestClassifier()))
models.append(('LOG', LogisticRegression()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('SVM', SVC()))
models.append(('DEC', DecisionTreeClassifier()))

for model_name, model in models:
    kfold = KFold(n_splits = 10, random_state=4)
    cv_results = cross_val_score(model,x_train, y_train, scoring = 'roc_auc', cv= kfold)
    final = "%s : %f(%f)" %(model_name, cv_results.mean(), cv_results.std())
    print(final)
```

```python
def performance(model):
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    print(accuracy_score(pred, y_test))
    print(classification_report(y_test, pred))
    print(confusion_matrix(y_test, pred))
  
```

```python
print(performance(DecisionTreeClassifier()))
```

### Using the Already explained Ensemble Methods above

#### - Max Voting in Python 

```python
#Using the Normal Pythonic implementation
#Using the max voting method on the DecisionTree, KNN, SVM and LogisticRegression models

model1 = DecisionTreeClassifier()
model1.fit(x_train, y_train)
model2 = KNeighborsClassifier()
model2.fit(x_train, y_train)
model3= LogisticRegression()
model3.fit(x_train, y_train)
model4 = SVC()
model4.fit(x_train, y_train)
model5 = RandomForestClassifier()
model5.fit(x_train, y_train)


pred1 = model1.predict(x_test)
pred2 = model2.predict(x_test)
pred3 = model3.predict(x_test)
pred4 = model4.predict(x_test)
pred5 = model5.predict(x_test)

from statistics import mode
ensembled_pred = np.array([])

for i in range(0, len(x_test)):
    ensembled_pred = np.append(ensembled_pred, mode([pred1[i], pred2[i], pred3[i], pred4[i], pred5[i]]))
    
print(accuracy_score(ensembled_pred, y_test))
print(classification_report(ensembled_pred, y_test))
    


```

```python
#Alternative Implementation using VotingClassifier in scikit-learn
from sklearn.ensemble import VotingClassifier
model1 = LogisticRegression()
model2 = DecisionTreeClassifier()
model3 = RandomForestClassifier()
model = VotingClassifier(estimators=[('log', model1), ('tree', model2), ('forest', model3)], voting='hard')
model.fit(x_train,y_train)
model.score(x_test,y_test)
print(classification_report(model.predict(x_test), y_test))

print(confusion_matrix(model.predict(x_test), y_test))
```

#### - Averaging Method

```python
#Pythonic Implementation

model1 = DecisionTreeClassifier()
model1.fit(x_train, y_train)
model2 = KNeighborsClassifier()
model2.fit(x_train, y_train)
model3= LogisticRegression()
model3.fit(x_train, y_train)
model4 = SVC()
model4.fit(x_train, y_train)
model5 = RandomForestClassifier()
model5.fit(x_train, y_train)


pred1 = model1.predict(x_test)
pred2 = model2.predict(x_test)
pred3 = model3.predict(x_test)
pred4 = model4.predict(x_test)
pred5 = model5.predict(x_test)

pred = np.round((pred1 + pred2 + pred3 + pred4 + pred5 )/5)

print(accuracy_score(pred, y_test))

print(classification_report(pred, y_test))



```

#### - Weighted Average Method




```python
model1 = DecisionTreeClassifier()
model1.fit(x_train, y_train)
model2 = KNeighborsClassifier()
model2.fit(x_train, y_train)
model3= LogisticRegression()
model3.fit(x_train, y_train)
model4 = SVC()
model4.fit(x_train, y_train)
model5 = RandomForestClassifier()
model5.fit(x_train, y_train)


pred1 = model1.predict(x_test)
pred2 = model2.predict(x_test)
pred3 = model3.predict(x_test)
pred4 = model4.predict(x_test)
pred5 = model5.predict(x_test)



pred = np.round(pred1*0.15 + pred2*0.1+ pred3*0.6+ pred4*0.1 + pred5*0.05)
print(accuracy_score(y_test,pred))

print(classification_report(pred, y_test))
```

#### - Stacking

```python
from mlxtend.classifier import StackingClassifier
#declaring the base learners - clf1, clf2, clf3
clf1 = KNeighborsClassifier(n_neighbors= 100)
clf2 = LogisticRegression()
clf3 = RandomForestClassifier(n_estimators = 170)


sclf = StackingClassifier(classifiers =[clf1, clf2, clf3, clf4], meta_classifier= clf1)
print(performance(sclf))
```

Randomized search CV - using the Random Forest Model

```python
from sklearn.model_selection import RandomizedSearchCV
#Performing the RandomizedSearchCV on the RandomForest Model
from sklearn.ensemble import RandomForestClassifier
#create the random grid containing the subset of hyperparameters 
random_grid = {
    "n_estimators": range(100,500,100),
    "max_depth": range(1,10,2),
    
    "max_features": range(10, 50,10),
    "min_samples_split": range(2,6,2),
    "bootstrap": [True, False]
}

print(random_grid)
```

#### Using the randomForest to search for the set of optimized hyperparameters randomly
#### A unique variable in the RandomizedSearchCV is the n_iter variable which specifies the number of different combinations to search through

```python
random_rf = RandomizedSearchCV(estimator= RandomForestClassifier(), param_distributions = random_grid, n_iter= 20, cv= 3, n_jobs=-1, verbose = 2)
```

```python
random_rf.fit(x_train, y_train)
```

```python
random_rf.best_score_
```

```python
random_rf.best_params_
```

```python
print(classification_report(random_rf.predict(x_test), y_test))
```
